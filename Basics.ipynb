````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
N = 10000 # We will have 10000 plots in my plot

# Initialize x and y vectors
sx = np.zeros(N)
sy = np.zeros(N)

for i in range(1,N):
  #Generate a random number {1,2,3}
  k = np.random.randint(1,4)

  # Update x and y points
  sx[i] = 0.5*sx[i-1] + k-1
  sy[i] = 0.5*sy[i-1]

  if k == 2:
    sy[i] += 2

plt.plot(sx,sy,'k.', markersize=1)
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()

# This plots the Sierpinski triangle -> Fractals -> Keep on zooming in
# and you will see more traingles
````
