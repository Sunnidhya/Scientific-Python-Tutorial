```python
# Alternative function
def fibSeq(n):
    v = [0,1,1]
    for i in range(n-3):
        v.append(sum(v[-2:]))
    print(v)
```

```python
# Exercise : make a Poisson like counter

# parameter lambda. positive number
# start at 1, multiply by random number uniform([0,1])
# target number is smaller than exp(-lambda)
# count the number of iterations until target<exp(-lambda)
```

```python
import math
import random

def PoissonCounter(n):
    # initialization
    target = 1
    count = 0

    # run algorithm
    while target>=math.exp(-n):
        target *= random.random()
        print(target)
        count += 1
    
    # result
    return count
```

```python
print(PoissonCounter(10))
```

```python
import numpy as np

p = np.zeros(100)

for i in range(len(p)):
    p[i] = PoissonCounter(10)


p
```
